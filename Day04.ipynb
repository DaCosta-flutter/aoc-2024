```kotlin
val example ="""MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX""".split("\n")
example
```

```kotlin
import utils.geometry.Direction
import utils.geometry.Point
import utils.geometry.moveTo
import utils.geometry.toGrid

var numXmas = 0

val grid = input.toGrid()
grid.forEach { point, _ ->
    numXmas += Direction.entries.count { point.isXmas(0, it) }
}

fun Point.isXmas(idx: Int, direction: Direction): Boolean {
    if (grid[this] != "XMAS"[idx]) {
        return false
    }
    return if (idx < "XMAS".lastIndex) this.moveTo(direction).isXmas(idx + 1, direction) else true
}

fun Point.isMas(): Boolean {
    if (grid[this] != 'A') {
        return false
    }
    val upLeft = grid[this.moveTo(Direction.UP_LEFT)]
    val upRight = grid[this.moveTo(Direction.UP_RIGHT)]
    val downLeft = grid[this.moveTo(Direction.DOWN_LEFT)]
    val downRight = grid[this.moveTo(Direction.DOWN_RIGHT)]

    if (!((upLeft == 'M' && downRight == 'S') || (upLeft == 'S' && downRight == 'M'))) {
        return false
    }
    if (!((downLeft == 'M' && upRight == 'S') || (downLeft == 'S' && upRight == 'M'))) {
        return false
    }
    return true
}

var numMas = 0

grid.forEach { point, _ ->
    numMas += if (point.isMas()) 1 else 0
}
numMas
```
